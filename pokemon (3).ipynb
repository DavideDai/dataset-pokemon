## outliers

```python
mean_value = df['total'].mean()
std_dev = df['total'].std()
std_dev
```

```python
outliers = df[(df['total'] > mean_value + 3 * std_dev) | (df['total'] < mean_value - 3 * std_dev)]
outliers
```

```python
plt.scatter(df.index, df['total'], label='punti totali pokemon')

# Evidenzia gli outliers nel grafico con un colore diverso
plt.scatter(outliers.index, outliers['total'], color='red', label='Outliers')

# Aggiungi la media e la deviazione standard al grafico
plt.axhline(y=mean_value, color='green', linestyle='-', label='Media')
plt.axhline(y=mean_value + 3 * std_dev, color='orange', linestyle='-', label='Â±3 Deviazioni Standard')
plt.axhline(y=mean_value - 3 * std_dev, color='orange', linestyle='-')

# Aggiungi etichette e legenda al grafico
plt.xlabel('Indice')
plt.ylabel('punti totali')
plt.title('Outliers dei punti totali')
plt.legend()

# Mostra il grafico
plt.show()
```

## splitting dataset

```python
from sklearn.model_selection import train_test_split

# Creare dati casuali per altezze (variabile indipendente) e pesi (variabile dipendente)
np.random.seed(0)
height = np.random.normal(160, 10, 100)
weight = 0.5 * height + np.random.normal(0, 5, 100)

# Suddividere il dataset in training set (70%) e test set (30%)
X_train, X_test, y_train, y_test = train_test_split(height, weight, test_size=0.3, random_state=42)

# Stampare le dimensioni dei training set e test set
print("Dimensioni del Training Set (altezze e pesi):", X_train.shape, y_train.shape)
print("Dimensioni del Test Set (altezze e pesi):", X_test.shape, y_test.shape)

# Creazione di un grafico a dispersione
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, label='Training Set', marker='*', color='red', alpha=0.7)
plt.scatter(X_test, y_test, label='Test Set', marker='*', color='blue', alpha=0.7)
plt.xlabel('altezza')
plt.ylabel('peso')
plt.title('Relazione tra Visite al Sito e Importo delle Vendite')
plt.legend()
plt.grid(True)
plt.show()

# Stampare le dimensioni dei training set e test set
print("Dimensioni del Training Set (Altezza e peso):", X_train.shape, y_train.shape)
print("Dimensioni del Test Set (Altezza e peso):", X_test.shape, y_test.shape)

```

```python
rf = RandomForestClassifier(n_estimators=40)
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
