```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv('pokemons (1).csv')

# Scatter plot
sns.scatterplot(x='height', y='weight', data=df)
plt.xlabel('height')
plt.ylabel('weight')
plt.show()
```

```python
df.shape
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
mean_value = df['total'].mean()
std_dev = df['total'].std()
std_dev
```

```python
outliers = df[(df['total'] > mean_value + 3 * std_dev) | (df['total'] < mean_value - 3 * std_dev)]
outliers
```

```python
plt.scatter(df.index, df['total'], label='punti totali pokemon')

# Evidenzia gli outliers nel grafico con un colore diverso
plt.scatter(outliers.index, outliers['total'], color='red', label='Outliers')

# Aggiungi la media e la deviazione standard al grafico
plt.axhline(y=mean_value, color='green', linestyle='-', label='Media')
plt.axhline(y=mean_value + 3 * std_dev, color='orange', linestyle='-', label='Â±3 Deviazioni Standard')
plt.axhline(y=mean_value - 3 * std_dev, color='orange', linestyle='-')

# Aggiungi etichette e legenda al grafico
plt.xlabel('Indice')
plt.ylabel('punti totali')
plt.title('Outliers dei punti totali')
plt.legend()

# Mostra il grafico
plt.show()
```
